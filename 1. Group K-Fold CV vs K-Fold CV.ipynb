# 3. Accuracy Scores Using:

## *Method1: K-Fold Cross Validation
## *Method2: Group K-Fold Cross Validation

```python
features = data.drop(['Plant ID','Treatment'], axis = 1)
target = data["Treatment"]
print("The unique number of plant ids are {}".format(data['Plant ID'].nunique()))
groups = list(data["Plant ID"])
```

## Method 1: "Group K-fold CV" no overlapping but samples sequence is shuffled. 

```python
gkf = GroupKFold(n_splits=63)
gkf
```

```python
cnt = 1
for train_index , test_index in gkf.split(features, target, groups=groups):
    print("\n Train Indeces:", train_index , "\n \n  Validation Indeces:",test_index)
    print("\n")
    print(f'Fold:{cnt}, Train set samples: {len(train_index)}, Test set samples:{len(test_index)}')
    cnt += 1
```

## Method 2: "K-fold CV" no overlapping and samples sequence is not shuffled. 

```python
kf = KFold(n_splits=63)
kf
```

```python
cnt = 1
for train_index , test_index in kf.split(features, target):
    print("\n Train Indeces:", train_index , "\n \n  Validation Indeces:",test_index)
    print("\n")
    print(f'Fold:{cnt}, Train set samples: {len(train_index)}, Test set samples:{len(test_index)}')
    cnt += 1
```

## 3.1 Create a function and a for loop to get the accuracy scores through all the K-Fold CV itterations.

```python
def get_score(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    return model.score(X_test, y_test)
```

```python
scores_ab = [] # Accuracy Scores of AdaBoost Classifier
scores_dt = [] # Accuracy Scores of Decision Trees Classifier
scores_rf = [] # Accuracy Scores of Random Forest Classifier
scores_lr = [] # Accuracy Scores of Logistic Regression Classifier

for train_index , test_index in kf.split(np.array(features), np.array(target)):
    X_train, X_test, y_train, y_test =  np.array(features)[train_index] , np.array(features)[test_index] , np.array(target)[train_index], np.array(target)[test_index]
    
    scores_ab.append(get_score(AdaBoostClassifier(), X_train, X_test, y_train, y_test))
    scores_rf.append(get_score(RandomForestClassifier(), X_train, X_test, y_train, y_test))
    scores_dt.append(get_score(DecisionTreeClassifier(), X_train, X_test, y_train, y_test))
    scores_lr.append(get_score(LogisticRegression(solver='liblinear',multi_class='auto'), X_train, X_test, y_train, y_test))
    
    
    
```

## 3.2 The mean accuracy scores through all the itterations.

```python
print("The avarage accuracy of AdaBoost is :", np.mean(scores_ab))
print("The avarage accuracy of Random Forest is :", np.mean(scores_rf))
print("The avarage accuracy of Decision Trees is :", np.mean(scores_dt))
print("The avarage accuracy of Logistic Regression is :", np.mean(scores_lr))
```
